```python
#Data Cleaning
#Here data is copied in player variable
player=data.copy()
```

```python
#Converting string data of value and wages column to float
def str2float(euros):
    if euros[-1]=='M':
        return float(euros[1:-1])*1000000
    elif euros[-1]=="K":
        return float(euros[1:-1])*1000
    else:
        return float(euros[1:])
    
```

```python
#he lambda keyword is used to define an anonymous function
player['Value']=player['Value'].apply(lambda x:str2float(x))
player['Wage']=player['Wage'].apply(lambda x:str2float(x))
```

```python
#Sorting overall rating on the basis of wage
#Displaying top 10 player
player.sort_values(by='Wage',ascending=False)[['Name','Value','Wage','Overall']][:10]
#Here we can see that L. Messi is in top list on basis of value
```

```python
#We can see that Real Madrid club has highest wage and overall rating
avgwageoverall = player.groupby("Club", as_index=False)["Wage","Overall"].mean()
avgwageoverall.sort_values(by="Wage",inplace=True,ascending=False)
avgwageoverall.head()
```
